```python
import os
```

```python
def delete_files_longer_than(folder_path, max_length):
    # Verifica che la cartella esista
    if not os.path.exists(folder_path):
        print(f"La cartella {folder_path} non esiste.")
        return

    # Elencare tutti i file nella cartella
    files = os.listdir(folder_path)

    # Iterare sui file e eliminare quelli con un nome piÃ¹ lungo della lunghezza massima
    for file_name in files:
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path) and len(file_name) > max_length:
            try:
                os.remove(file_path)
                print(f"Eliminato: {file_name}")
            except Exception as e:
                print(f"Errore durante l'eliminazione di {file_name}: {e}")
```

```python
cartella1 = "C:/Users/Paolo/Desktop/Prova2/Traffic Sign Detenction/Traffic Sign Detenction/test/images"
cartella2 = "C:/Users/Paolo/Desktop/Prova2/Traffic Sign Detenction/Traffic Sign Detenction/test/labels"
lunghezza_massima = 10

# Esegui la funzione per eliminare i file
delete_files_longer_than(cartella1, lunghezza_massima)
delete_files_longer_than(cartella2, lunghezza_massima)
```
