```python
from ultralytics import YOLO
import cv2
import os
```

```python
# Load a pretrained YOLOv8n model
model = YOLO('../../models/detenction/train_7_classes_split_70_15_15/weights/best.pt')
```

```python
# Define path to video file
#source = 'C:/Users/Paolo/Downloads/Telegram Desktop/20231128_131732_R.mp4'
source = 'C:/Users/Paolo/Downloads/Telegram Desktop/20231128_132114.mp4'
#source = 'C:/Users/Paolo/Downloads/Telegram Desktop/20231128_131732_R.mp4'
#source = 'C:/Users/Paolo/Downloads/Telegram Desktop/20231128_131732_R.mp4'
```

```python
# Run inference on the source
# results = model(source, stream=True)  # generator of Results objects
```

```python
# Set a counter to keep track of frames
frame_counter = 0
# List to store annotated frames
annotated_frames = []
cap = cv2.VideoCapture(source)

while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()
    if success:
        # Increment the frame counter
        frame_counter += 1
        # Process and display every other frame
        if frame_counter % 2 == 0:
            # Run YOLOv8 inference on the frame
            results = model(frame)
            # Visualize the results on the frame
            annotated_frame = results[0].plot()
            # Append annotated frame to the list
            annotated_frames.append(annotated_frame)
            # Display the annotated frame (optional, for real-time viewing)
            cv2.imshow("YOLOv8 Inference", annotated_frame)
        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break
# Release the video capture object
cap.release()
# Check if there are any frames to process
if annotated_frames:
    # Create a VideoWriter object
    output_video_path = "./videos/" + os.path.basename(source) + ".avi"
    print(output_video_path)
    fourcc = cv2.VideoWriter_fourcc(*"XVID")
    output_video = cv2.VideoWriter(output_video_path, fourcc, 30.0, annotated_frames[0].shape[1::-1])
    # Write annotated frames to the video file
    for annotated_frame in annotated_frames:
        output_video.write(annotated_frame)
    # Release the VideoWriter object
    output_video.release()
# Close the display window
cv2.destroyAllWindows()
```
